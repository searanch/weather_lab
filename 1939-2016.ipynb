```python
# load nessacary packages

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# load file

file = "../data/MN1938_2019.csv"

df = pd.read_csv(file)
df.head()
df.columns
```

```python
# mask
start_date = '1939-01'
end_date = '1990-12'

#set a mask to apply the df filter
mask = (df['DATE'] >= start_date) & (df['DATE'] <= end_date)

# Asign mask to df
df = df.loc[mask]
df.head()
```

```python
df.head()
df.tail()
```

```python
LOW = df["TMIN"]
#HIGH =df["TMAX"]

plt.figure(figsize=(18,8))


plt.xlabel(" Minneapolis")
plt.ylabel("Farenheit")

# giving a handle for lows
lows_handle, = plt.plot( LOW, label="LOW")
# giving a handle for highs
#highs_handle, = plt.plot( HIGH, label="HIGH")

plt.legend(loc="best")

# Draw a horizontal line with 0.25 transparency to represent above/below zero marker
plt.hlines(0, 600, 0, alpha=0.40)

plt.title("Average Low Tempature 1938-1990 in Minneapolis")
```
