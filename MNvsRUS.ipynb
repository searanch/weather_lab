```python
#import neccacary packages
import pandas as pd

import numpy as np
import matplotlib.pyplot as plt
import matplotlib.ticker as ticker
```

```python
# load dataset

file = "../data/Russia96_98.csv"

df = pd.read_csv(file)
df.head()
df.tail()
```

```python
# load Minn. data set

file= "../data/Minneapolis96_98.csv"

df_MN = pd.read_csv(file)

df_MN.head()
df_MN.tail()
```

```python
# mask
start_date = '1996/12/31'
end_date = '1998/3/01'

#set a mask to apply the df filter
mask = (df_MN['DATE'] >= start_date) & (df_MN['DATE'] <= end_date)

# Asign mask to df
df_MN = df_MN.loc[mask]
df_MN.head()
df_MN.tail()
```

```python
# mask
start_date = '1996/12/31'
end_date = '1998/3/01'

#set a mask to apply the df filter
mask = (df['DATE'] >= start_date) & (df['DATE'] <= end_date)

```

```python
# Asign mask to df
df = df.loc[mask]
df.head()
df.tail()
```

```python
# merge df and df_MN

combo = pd.merge(df, df_MN, on= "DATE")
combo.head()
combo.tail()

```

```python
reduced = combo[['NAME_x', 'DATE',"TMAX_x",'TMIN_x','NAME_y','TMAX_y','TMAX_x']].copy()
reduced
```

```python
reduced.columns = ['Russia','Date','Russia High','Russia Low','Minneapolis','MN High','MN Low']
reduced.head()
```

```python
# plotting


Russia = reduced['Russia Low']
Minneapolis = reduced['MN Low']
plt.figure(figsize=(18,8))




plt.xlabel("Was it colder in Russia or Minneapolis?")
plt.ylabel("Farenheit")

# giving a handle for lows
lows_handle, = plt.plot( Russia, label="Russia")
# giving a handle for highs
highs_handle, = plt.plot( Minneapolis, label="Minneapolis")
plt.legend(loc="best")

# Draw a horizontal line with 0.25 transparency to represent above/below zero marker
plt.hlines(0, 0, 515, alpha=0.40)

plt.title("Low Tempature December 1996-April 1998")

```
